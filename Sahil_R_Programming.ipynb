```r
# bill

bill <- function(price,quantity,gst){
  total_amount <- price*quantity
  gst_amount <- total_amount*gst/100
  total_price <- total_amount+gst_amount

  cat("\n--------bill--------\n")
  cat("\nAmount = ",price,"/-")
  cat("\nquantity = ",price,"/-")
  cat("\nGST = ",gst,"%")
  cat("\nGST Amount  = ",gst_amount,"/-")
  cat("\nTotal Price (With GST)  = ",total_price,"/-")

}
```

```r
bill (578,5,12)
```
