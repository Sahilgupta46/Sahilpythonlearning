```python
# EXAMPLE :

data = {'name': 'joy', 'age': 25, 'city': 'Dhaka'}
print(data)
```

```python
# access the items of a dictionary
print(data['name'])
print(data['age'])
```

```python
# get all keys of a dictionary
print(data.keys())
```

```python
# get all values of a dictionary
print(data.values())
```

```python
# get data from keys
print(data.get('name'))
```

```python
# get all data from using items() method
print(data.items())
```

```python
# update values of a dictionary
data['age'] =30
data['city'] = 'Lucknow'
print(data)
```

```python
# add new key-value pair in a dictionary
data['country'] = 'Uttar Pardesh'
print(data)
```

```python
# delelte a key-value pair from a dictionary
del data['county']
print(data)
```

```python
data['country'] = 'India'
print(data)
```

```python
# iterate through a dictionary
# fetch data from loop
for i in data:
   print(i,'=', data[i])
```

```python
# fetch data from items() method
for key, value in data.items():
   print(key, '=', value)
```

```python
# check if a key exists in a dictionary

print('name'in data)
print('country' in data)
print('city' in data)
```

```python
# how to take user input and store in a dictionary

# step 1 : create a dictionary
# step 2 : take user input
# step 3 : store the input in the dictionary
# step 4 : print the dictionary

data = {}

name = input('Enter your name :')
age = input('Enter your age :')
city = input('Enter your city :')

data['name'] = name
data['age'] = age
data['city'] = city

print(data)
```
